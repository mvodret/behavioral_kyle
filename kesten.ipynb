```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from math import gamma
from scipy.stats import lognorm,invgamma
import pandas as pd
import scipy.stats as stats
import matplotlib as mpl
```

```python
mu = 1
n = 100000
m = 10000
```

```python
%%time

x = np.ones((n,m))
for i in range(1,n):
    x[i] = 0.5+0.5*np.random.exponential(mu,m)*x[i-1]
```

```python
plt.hist(x.flatten(),bins = 1000,density = True)
plt.plot(np.arange(1,10**6),0.8*np.arange(1,10**6)**(-3.45))

plt.legend()
plt.yscale('log')
plt.xscale('log')
plt.ylim(10**(-9),10)
plt.xlim(1,200)
plt.show()
plt.show()
```
